```python
import os
from pdf2image import convert_from_path, convert_from_bytes
import PIL as image

import tempfile
```

```python
path = "./pdfs"
dir_list = os.listdir(path)
```

```python
dir_list
```

```python
for index in range(len(dir_list)):
    i = len(dir_list) - (index+1)
    with tempfile.TemporaryDirectory() as path:
        images_from_path = convert_from_path(f"./pdfs/{dir_list[i]}", output_folder="./converted",fmt="jpeg")
```
