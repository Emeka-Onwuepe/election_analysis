```python
import json
import pandas as pd
import re
```

```python
base = './'

def read_json(name,key=None):
    name = f"{base}/{name}"
    try:
        with open(name,'r') as openfile:
            data = json.load(openfile)
            if key:
                return data[key]
            return data
    except (FileNotFoundError, KeyError)as e:
        if key:
            return []
        return {}

def write_json(data,name,key=None):
    name = f"{base}/{name}"
    if key:
        initial_data = read_json(name)
        initial_data[key] = data
        data = initial_data
    data = json.dumps(data)
    with open(name,'w') as outfile:
        outfile.write(data)


def backup_json(name):
    original = read_json(name)
    back_up =  f"{base}/{name}"
    write_json(original,back_up)
```

```python
data = read_json('presidential_2023-02-25_v3.json')
```

```python
rows = []
lg_wards_polling = data['lg_wards_polling']
for lg_name,lg_data in lg_wards_polling.items():
    *lg,state = lg_name.split(' - ')
    lg = ''.join(lg).split('(')
    for ward_name,ward_data in lg_data.items():
        if ward_name not in ['status','missing']:
            *ward,_ = ward_name.split(' - ')
            ward = "".join(ward).split('(')
            
            for polling_unit in ward_data:
                code,polling_name = polling_unit.split('--->')
                state_code,lg_code,ward_code,pu_code = code.split('/')
                info = [state.strip(),lg[0].strip(),ward[0].strip(),
                        state_code,lg_code,ward_code,pu_code.strip(),
                        code.strip(),polling_name.strip()]
                rows.append(info)
```

```python
columns = ['state','lg','ward','state_code','lg_code','ward_code','pu_code','code','location']
df = pd.DataFrame(rows,columns=columns)
```
